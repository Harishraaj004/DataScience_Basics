```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,56,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.xlabel('day')
plt.ylabel('temperatue')
plt.title('weather')
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')

plt.legend(loc='best', shadow=True, fontsize="small")
```

```python
plt.xlabel('day')
plt.ylabel('temperatue')
plt.title('weather')
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')

plt.legend(loc='best', shadow=True, fontsize="small")
plt.grid()
```

```python
import numpy as np
```

```python
company=['google','microsoft','amazon','meta']
revenue=[90,136,89,27]
```

```python
y= np.arange(len(company))
y
```

```python
plt.bar(company,revenue)
```

```python
profit=[40,2,34,12]
plt.xticks(y,company)
plt.ylabel('revenue in $')
plt.title('US stock values')
plt.bar(y,revenue, label= 'revenue')
plt.bar(y,profit, label= 'profit')
plt.legend()
```

```python
xpos= np.arange(len(company))
xpos
```

```python
profit=[40,2,34,12]
plt.xticks(y,company)
plt.ylabel('revenue in $')
plt.title('US stock values')
plt.bar(xpos-0.2,revenue, width=0.4, label= 'revenue')
plt.bar(xpos+0.2,profit, width=0.4, label= 'profit')
plt.legend()
```

```python
profit=[40,2,34,12]
plt.xticks(y,company)

plt.barh(xpos-0.2,revenue, label= 'revenue')
plt.barh(xpos+0.2,profit,  label= 'profit')
plt.legend()
```
